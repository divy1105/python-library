```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('winequality.csv')
df2=pd.read_csv('tips.csv')
df
```

```python
sns.countplot(x='type',data=df,hue='quality')
```

```python
sns.countplot(data=df2,x='day',hue='sex')
```

```python
# no of cutstomer visited based on gender on weekend
df2['weekend'] = df2['day'].isin(['Sat','Sun'])
sns.countplot(data=df2,x='weekend',hue='sex')
x = sns.countplot()
for i in x.containers:
    x.bar_label(i)
```

```python
# Barplot
# it will display the np.average values of a perticular numerical column with respective to categorical colomn
```

```python
# sns.barplot(x='categorical column',y='numerical column',dta=df)
```

```python
a= sns.barplot(x='type',y='pH',data=df)
for i in a.containers:
    a.bar_label(i)
```

```python
# average tip based on day
a=sns.barplot(x='day',y='tip',data=df2,hue='sex',errorbar=None)
a.legend(loc='upper right')
for i in a.containers:
    a.bar_label(i)
```
