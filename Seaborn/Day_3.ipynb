```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/Electric_Vehicle_Population_Data.csv')

# Display the first 5 rows
display(df.head())
```

```python
# Display the DataFrame's information to check data types and non-null values
df.info()
```

```python
# Identify and count missing values before cleaning
print("Missing values before cleaning:")
print(df.isnull().sum())

# Drop rows where 'Electric Range' or 'Base MSRP' are missing
df.dropna(subset=['Electric Range', 'Base MSRP'], inplace=True)

# Impute 'Legislative District' with the mode
mode_legislative_district = df['Legislative District'].mode()[0]
df['Legislative District'].fillna(mode_legislative_district, inplace=True)

# Fill 'Vehicle Location' missing values with 'Unknown', inplace=True
df['Vehicle Location'].fillna('Unknown', inplace=True)

# Drop any remaining rows with missing values
df.dropna(inplace=True)

# Display the DataFrame's information after cleaning to confirm no nulls
print("\nMissing values after cleaning:")
print(df.isnull().sum())

# Display the first 5 rows of the cleaned DataFrame
display(df.head())
```

```python
# Display the unique values in the 'Clean Alternative Fuel Vehicle (CAFV) Eligibility' column
display(df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].unique())
```

```python
# Rename the values in the 'Clean Alternative Fuel Vehicle (CAFV) Eligibility' column
df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] = df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].replace({
    'Clean Alternative Fuel Vehicle Eligible': 'Eligible',
    'Eligibility unknown as battery range has not been researched': 'Unknown',
    'Not eligible due to low battery range': 'Not eligible'
})

# Display the unique values after renaming to confirm
display(df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].unique())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for the plots
sns.set_style("whitegrid")

# Plot the distribution of 'Model Year'
plt.figure(figsize=(10, 6))
sns.histplot(df['Model Year'], bins=20, kde=True)
plt.title('Distribution of Model Year')
plt.xlabel('Model Year')
plt.ylabel('Count')
plt.show()

# Plot the top 10 'Make' of electric vehicles
plt.figure(figsize=(12, 7))
df['Make'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Electric Vehicle Makes')
plt.xlabel('Make')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plot the distribution of 'Electric Range'
plt.figure(figsize=(10, 6))
sns.histplot(df['Electric Range'], bins=30, kde=True)
plt.title('Distribution of Electric Range')
plt.xlabel('Electric Range (miles)')
plt.ylabel('Count')
plt.show()
```

```python
sns.lineplot(x='Model Year', y='Electric Range', data=df)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a relplot to visualize the relationship between 'Base MSRP' and 'Electric Range'
plt.figure(figsize=(12, 7))
sns.relplot(x='Base MSRP', y='Electric Range', data=df, kind='scatter', height=6, aspect=1.5)
plt.title('Relationship between Base MSRP and Electric Range')
plt.xlabel('Base MSRP ($)')
plt.ylabel('Electric Range (miles)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Aggregate data to calculate the mean Electric Range for each Model Year and Electric Vehicle Type
agg_df = df.groupby(['Model Year', 'Electric Vehicle Type'])['Electric Range'].mean().reset_index()

# Create the relplot similar to the example
g = sns.relplot(
    data=agg_df,
    x="Model Year",
    y="Electric Range",
    col="Electric Vehicle Type", # Facet the plot by Electric Vehicle Type
    hue="Electric Vehicle Type", # Color the lines by Electric Vehicle Type
    kind="line",
    palette="crest",
    linewidth=4,
    zorder=5,
    col_wrap=2, # Since there are two Electric Vehicle Types
    height=4,
    aspect=1.5,
    legend=False, # Following the example's legend setting
)

# Set titles and labels for better readability
g.set_axis_labels("Model Year", "Mean Electric Range (miles)")
g.set_titles("{col_name}")
plt.suptitle('Mean Electric Range Trend by Model Year and Vehicle Type', y=1.02, fontsize=16)
g.tight_layout()
plt.show()
```
