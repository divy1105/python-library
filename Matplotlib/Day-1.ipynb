```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df=pd.read_csv('sales_dataset.csv')
```

```python
a=np.arange(1,11)
b=a**2
c=a**3
```

```python
plt.figure(figsize=(14,10),facecolor='magenta')
plt.plot(a,b,label='square',color='red',linewidth=2,linestyle='dashed',marker='.',markerfacecolor='black',markeredgecolor='black')
plt.plot(a,c,label='cube',color='purple',linestyle='dashdot',marker='.',markerfacecolor='black',markeredgecolor='black')
plt.xlabel('values',color='#522626',fontstyle='italic')
plt.ylabel('square',color='#1B1C69',fontstyle='italic')
plt.title('Graph')
plt.legend(title='Index',facecolor='skyblue',loc='best',labelcolor='#675382',fontsize=15,title_fontsize=15)
for i,j in  zip(a,c):
  plt.annotate(f'{j}',(i,j),ha='right')
for i,j in  zip(a,b):
  plt.annotate(f'{j}',(i,j),ha='left')
plt.grid()
plt.show()
```

```python
# total sales based on region
df.groupby('Region')['TotalSales'].sum()
```

```python
plt.figure(figsize=(10, 6),facecolor='lightgreen')
sales_by_region = df.groupby('Region')['TotalSales'].sum()
sales_by_region = pd.DataFrame(sales_by_region)
sales_by_region = sales_by_region.reset_index() # Assign the result of reset_index()
sales_by_region.plot(x='Region', y='TotalSales', kind='line', color='purple') # Specify x and y for accurate plotting
# Use iterrows to get numerical index for annotation
for index, row in sales_by_region.iterrows():
  plt.annotate(f'{row["TotalSales"]:.2f}', (index, row['TotalSales']), ha='right')
plt.title('Total Sales by Region',color='red')
plt.xlabel('Region',color='blue')
plt.ylabel('Total Sales',color='darkblue')
plt.grid()
plt.show()
```
