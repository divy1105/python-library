### 1.Renaming the columns 

In the case of Rename of columns, we will create a dictionary which will be consisting of old columnName as the key and the new columnName as the value.

* Syntax ----> 
  df.rename(columns=dictionary,inplace=True)

```python
old=list(df.columns)
new=['Pclass', 'Name', 'Gender', 'Age', 'Sibsp', 'Parch', 'Ticket', 'Fare',
       'Cabin', 'Embarked', 'Survived']

new_columns={}
for i,j in zip(old,new):
    new_columns[i]=j
```

```python
df.rename(columns=new_columns,inplace=True)
```
