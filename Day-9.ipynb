```python
import pandas as pd
df=pd.read_csv("zomato_cleaned_data.csv")
df
```

```python
df.columns
```

-------------------------------------------------------------------------------
1. Create a pivot table to find the count of restaurants in 
each location. 

```python
a=df.pivot_table(index='location',values='rest_type',aggfunc='count')
a
```

-------------------------------------------------------------------------------
2.Find the average rate for each rest_type.

```python
b=df.pivot_table(index='rest_type',values='rate',aggfunc='mean')
b
```

--------------------------------------------------------------------------------
3. Create a pivot table showing the average cost for 2 
across different location and type combinations.

```python
c=df.pivot_table(index=['location','type'],values='costfor2',aggfunc='mean')
c
```

--------------------------------------------------------------------
4. Find the total votes received by each rest_type for each type (Buffet, Cafes, Delivery, etc.).

```python
d=df.pivot_table(index='rest_type',columns='type',values='votes',aggfunc='sum')
d
```

---------------------------------------------------------------------------------
5. Count how many restaurants accept online orders 
(online_order) across each location. 

```python
data=df[df['online_order']=='Yes']
data
data.pivot_table(index='location',values='name',aggfunc='count')
```

------------------------------------------------------------------------------------
6. Create a pivot table showing the maximum rate for each combination of location and rest_type. 

```python
f = df.pivot_table(values='rate',index=['location','rest_type'],aggfunc='max')
f
```
