```python
import pandas as pd
```

```python
data = pd.read_csv("hr_data.csv")
data
```

```python
sales_left = data[(data['department'] == 'sales') & (data['left'] == 1)]
sales_left
```

# distribution of sal based on department

```python
data.pivot_table(values=['salary'],
                index=['department'],
                aggfunc=['count'])
```

```python
data.pivot_table(index='department',aggfunc='count')
```

```python
data.pivot_table(index='department',aggfunc='sum')
```

```python
# we can use more than one functions in aggfunc
data.pivot_table(values='number_project',index='department',aggfunc=['count','sum'])
```

```python
#avg monthly hours based on department
data.pivot_table(values='average_montly_hours',index='department',aggfunc='mean')
```

```python
# promotion last 5 year based sal
data.pivot_table(values='promotion_last_5years',index='salary',aggfunc='count')
```
