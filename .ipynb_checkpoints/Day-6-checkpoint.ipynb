# Obj columns ----->

* In the case of obj columns if there are less nmber of null value that is below 5% it can replace with frequently occurring value


* If more then 70% of data is null then the column is recommended to be drop

* If the null value are between 20-50% & the column is a categorial column, we can replace those  value is unknown

```python
# Define valid categories
valid_categories = ["Unfurnished", "Semi-Furnished", "Furnished"]

# Apply function to clean the column
df["furnishing"] = df["furnishing"].apply(
    lambda x: x if x in valid_categories else "Unknown"
)

# Check the cleaned value counts
print(df["furnishing"].value_counts())
```

```python
df['furnishing'].unique()
```

```python
# Define valid categories
valid_categories = ["Unfurnished", "Semi-Furnished", "Furnished"]

# Replace all other values with "Unknown" directly in the original DataFrame
df["furnishing"].replace(
    to_replace=[x for x in df["furnishing"].unique() if x not in valid_categories],
    value="Unknown",
    inplace=True
)

# Verify the changes
print(df["furnishing"].value_counts())
```

# Numerical columns

* In the case of numerical columns if there are less number of null value that is below 5% it can replace with Median


* And the null value are more than 70% the column is recommended to be dropped

* There are other 2 methods which are use to fill the null values in numerical coulmns butIt should be onlu use one a data in which the values are continuous 

1. Forward fill denoted by f-fill
2. Backward fill denoted by b-fill

* and also it is use in date columns

```python
# create a dataframe 4 rows - abcdef marks - 40,55,80,90
import numpy as np
data = {'name' : ['a','b','c','d','aa','bb','d','e','f'],
         'marks' : [40,55,80,90,np.NaN,np.NaN,45,np.NaN,10]}
data = pd.DataFrame(data)
data
```

```python
data['marks'].median()
data.fillna(data['marks'].mean().round())
```
