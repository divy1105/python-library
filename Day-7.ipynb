```python
df.dropna(inplace=True)
```

```python
df.rename(columns={'numero': 'Fires','ano': 'year','mes': 'month','estado': 'city','encontro': 'date'}, inplace=True)
df
```

```python
# Step 1: Split and keep only the number part
df['Fires'] = df['Fires'].str.split().str[0]

# Step 2: Convert to numeric (floats if needed)
df['Fires'] = pd.to_numeric(df['Fires'], errors='coerce')

# Step 3: Handle decimals before casting
# Option A: Round to nearest integer
df['Fires'] = df['Fires'].round().astype('Int64')

# Option B: Floor (drop decimal part)
# df['Fires'] = df['Fires'].astype(int)
```

```python
# Define the mapping dictionary
month_map = {
    'Janeiro': 'January',
    'Fevereiro': 'February',
    'Março': 'March',
    'Abril': 'April',
    'Maio': 'May',
    'Junho': 'June',
    'Julho': 'July',
    'Agosto': 'August',
    'Setembro': 'September',
    'Outubro': 'October',
    'Novembro': 'November',
    'Dezembro': 'December'
}

# Option 1: Explicit reassignment (recommended)
df['month'] = df['month'].replace(month_map)

# Option 2: Whole DataFrame replace with dict (safe inplace)
df.replace({'month': month_map}, inplace=True)
```

# Groupby

* Groupby is used to group the data of the perticular column

* Syntax -----> df.groupby(columnName)

```python
a=df.groupby('month')
a.get_group('January')
```

```python
a=df.groupby('month').count()
a

```

```python
data=df.groupby('month')
jan=data.get_group('January')
jan['Fires'].sum()
```

```python
data=df.groupby('city')
jan=data.get_group('Rio')
jan['Fires'].sum()
```
