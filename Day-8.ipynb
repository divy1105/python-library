# Pivot Table

* In the case of groupby function we can only apply single aggregate function but using pivot table we can apply multiple aggregate functionon the column
* Syntax ---->  df.pivot_table() 

```python
import pandas as pd
data=pd.read_csv("titanic.csv") 
data
```

```python
data.pivot_table(values=['survived'],
                index=['sex'],
                aggfunc=['sum'])
```

```python
data.pivot_table(values=['age'],
                index=['sex'],
                aggfunc=['mean'])
```

```python
data.pivot_table(values=['sex'],
                index=['survived'],
                aggfunc=['count'])
```

```python
data.pivot_table(values=['fare'],
                index=['pclass'],
                aggfunc=['sum'])
```
